```python
import os

def print_labels_more_than_one(folder_labels):
    found = False

    for fname in os.listdir(folder_labels):
        if not fname.lower().endswith(".txt"):
            continue

        fpath = os.path.join(folder_labels, fname)
        if not os.path.isfile(fpath):
            continue

        with open(fpath, "r", encoding="utf-8") as f:
            lines = [line for line in f if line.strip()]

        if len(lines) > 1:
            print(f"{fname} -> {len(lines)} objek")
            found = True

    if not found:
        print("Tidak ada file dengan lebih dari 1 objek.")


# contoh pemakaian
labels_dir = r"C:\Users\USER\Documents\a skripsi\workv1\self-training\5-iter\pseudo\labels"
print_labels_more_than_one(labels_dir)
```
