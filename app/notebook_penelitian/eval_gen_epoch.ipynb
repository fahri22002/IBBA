```python
import os
import csv
from ultralytics import YOLO

# ===================== KONFIGURASI =====================
DATA_YAML = r"C:/Users/USER/Documents/a skripsi/riset/data_eval_final/test generalisasi/data.yaml"

MODEL_PATHS = [
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/1-iter/model1/weights/best.pt",
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/2-iter/model2/weights/best.pt",
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/3-iter/model3/weights/best.pt",
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/4-iter/model4/weights/best.pt",
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/5-iter/model5/weights/best.pt",
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/6-iter/model6/weights/best.pt",
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/7-iter/model7/weights/best.pt",
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/8-iter/model8/weights/best.pt",
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/9-iter/model9/weights/best.pt",
    r"C:/Users/USER/Documents/a skripsi/workv1/self-training/10-iter/model10/weights/best.pt",
]

OUTPUT_DIR = "./generalisasi_epoch-1"
CSV_OUT = os.path.join(OUTPUT_DIR, "results.csv")

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ===================== HEADER CSV =====================
header = [
    "iteration",
    "TP", "FP", "TN", "FN",
    "precision", "recall", "f1_score",
    "mAP50", "mAP50_95"
]

rows = []

# ===================== LOOP VALIDASI =====================
for idx, model_path in enumerate(MODEL_PATHS, start=1):
    print(f"\nValidating model iterasi {idx}")
    model = YOLO(model_path)

    metrics = model.val(
        data=DATA_YAML,
        imgsz=640,
        conf=0.5,
        # iou=0.5,
        project=OUTPUT_DIR,
        name=f"iterasi_{idx}",
        save_json=False,
        plots=True
    )

    # Confusion Matrix
    cm = metrics.confusion_matrix.matrix
    # Format:
    # [[TP, FP],
    #  [FN, TN(background)]]

    TP = int(cm[0, 0])
    FP = int(cm[0, 1])
    FN = int(cm[1, 0])
    TN = 0  # Tidak bermakna untuk object detection
    print(f"TP: {TP}, FP: {FP}, TN: {TN}, FN: {FN}")

    # precision = float(metrics.box.precision[0])
    # recall = float(metrics.box.recall[0])
    # f1 = 2 * precision * recall / (precision + recall + 1e-9)
    precision = float(metrics.box.p[0])
    recall = float(metrics.box.r[0])
    f1 = float(metrics.box.f1[0])


    mAP50 = float(metrics.box.map50)
    mAP5095 = float(metrics.box.map)

    rows.append([
        idx,
        TP, FP, TN, FN,
        precision, recall, f1,
        mAP50, mAP5095
    ])

# ===================== SIMPAN CSV =====================
with open(CSV_OUT, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(rows)

print("\n✅ Validation selesai")
print(f"Hasil disimpan ke: {CSV_OUT}")
```
